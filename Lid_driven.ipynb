Solve the lid-driven Navier-Stokes equation using the finite difference method for spatial discretization and employs the backward Euler time-stepping technique for time discretization, along with Chrin's projection method

Momentum equation: 
\begin{equation}
\frac{\partial \mathbf{u}}{\partial t} + (\mathbf{u} \cdot \nabla) \mathbf{u} = -\frac{1}{\rho} \nabla p + \nu \nabla^2 \mathbf{u} + \mathbf{f}
\end{equation}

Imcompressibility:
\begin{equation}
\nabla  \cdot \mathbf{u} = 0
\end{equation}

$\mathbf{u}$: Velocity (2d vector)<br>
$p$: Pressure<br>
$\mathbf{f}$: Force (set to be $\mathbf{0}$)<br>
$\nu$: Viscosity<br>
$\rho$: Density<br>
$t$: Time <br>

Lid-driven cavity


Projection method to solve Navier-Stokes equation:

\begin{align}
\rho\frac{u_{i,j}^{*}-u_{i,j}^{n}}{\Delta t}-\nu \left( \frac{u_{i-1,j}^{*}-2u_{i,j}^*+u_{i+1,j}^*}{\Delta x^2} + \frac{u_{i,j-1}^*-2u_{i,j}^*+u_{i,j+1}^*}{\Delta y^2}\right) + u^{n}_{i,j}\frac{u^n_{i,j}-u^n_{i-1,j}}{\Delta x} +  v^{n}_{i,j}\frac{u^n_{i,j}-u^n_{i,j-1}}{\Delta y}= 0,
\end{align}

\begin{align}
\rho\frac{v_{i,j}^{*}-v_{i,j}^{n}}{\Delta t}-\nu \left( \frac{v_{i-1,j}^{*}-2v_{i,j}^*+v_{i+1,j}^*}{\Delta x^2} + \frac{v_{i,j-1}^*-2v_{i,j}^*+v_{i,j+1}^*}{\Delta y^2}\right) + u^{n}_{i,j}\frac{v^n_{i,j}-v^n_{i-1,j}}{\Delta x} +  v^{n}_{i,j}\frac{v^n_{i,j}-v^n_{i,j-1}}{\Delta y}= 0,
\end{align}

\begin{align}
-\left( \frac{p_{i-1,j}^{n+1}-2p_{i,j}^{n+1}+p_{i+1,j}^{n+1}}{\Delta x^2} + \frac{p_{i,j-1}^{n+1}-2p_{i,j}^{n+1}+p_{i,j+1}^{n+1}}{\Delta y^2}\right) = -\frac{\rho}{\Delta t} \left( \frac{u^*_{i,j}-u^*_{i-1,j}}{\Delta x} +  \frac{v^*_{i,j}-v^*_{i,j-1}}{\Delta y}\right),
\end{align}

\begin{align}
u^{n+1}_{i,j} = u^{*}_{i,j} -\frac{\rho}{\Delta t}  \frac{p^{n+1}_{i,j}-p^{n+1}_{i-1,j}}{\Delta x},
\end{align}

\begin{align}
v^{n+1}_{i,j} = v^{*}_{i,j} -\frac{\rho}{\Delta t}  \frac{p^{n+1}_{i,j}-p^{n+1}_{i,j-1}}{\Delta y}.
\end{align}

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.sparse.linalg import spsolve
from scipy.sparse import diags, kron, eye, csr_matrix,coo_matrix, isspmatrix_csr,spdiags
import matplotlib.pyplot as plt
```

```python
lx = 1        # width of box
ly = 1        # height of box
nx = 50      # number of x-gridpoints
ny = 50       # number of y-gridpoints
mu = 0.01
x = np.linspace(0, lx, nx+1)
hx = lx / nx
y = np.linspace(0, ly, ny+1)
hy = ly / ny
xm = np.linspace(0.5*hx, lx-0.5*hx, nx)  # x-coordinate
ym = np.linspace(0.5*hy, ly-0.5*hy, ny)  # y-coordinate
X, Y = np.meshgrid(x, y)
dt = hx/8
t_start = 0
t_final = dt * 10000
```

```python
uN = x*0 + 1
uS = x*0
uE = ym*0
uW = ym*0

vN = xm*0
vS = xm*0
vE = y*0
vW = y*0

uW = uW.reshape((1,uW.shape[0]))
uE = uE.reshape((1,uE.shape[0]))
uS = uS.reshape((uS.shape[0],1))
uN = uN.reshape((uN.shape[0],1))

vS = vS.reshape((vS.shape[0],1))
vN = vN.reshape((vN.shape[0],1))
vW = vW.reshape((1,vW.shape[0]))
vE = vE.reshape((1,vE.shape[0]))

```

```python
Ubc = mu*(np.concatenate((2*uS[1:-1,:],np.zeros((nx-1,ny-2)),2*uN[1:-1,:]), axis=1)/hy**2 +\
      np.concatenate((uW,np.zeros((nx-3,ny)),uE), axis=0)/hx**2)



Vbc = mu*(np.concatenate((vS,np.zeros((nx,ny-3)),vN), axis=1)/hy**2 +\
      np.concatenate((2*vW[:,1:-1],np.zeros((nx-2,ny-1)),2*vE[:,1:-1]), axis=0)/hx**2)

divbc = (np.concatenate((-vS, np.zeros((nx, ny-2)), vN), axis=1) / hy +\
        np.concatenate((-uW, np.zeros((nx-2, ny)), uE), axis=0) / hx).flatten(order='F')
 
```

```python
def D(n, h, b):
    # Create the tridiagonal matrix data
    main_diag = 2*np.ones(n)
    off_diag = -np.ones(n-1)

    # Update first and last elements of the main diagonal
    main_diag[0] = main_diag[-1] = b

    # Create the sparse tridiagonal matrix
    L_sparse = diags([off_diag, main_diag, off_diag], [-1, 0, 1], shape=(n, n), format='csr')

    # Transpose and divide by h^2
    L = L_sparse.transpose() / h**2

    return L

```

```python
Lu = eye((nx-1)*ny)/(dt) + mu*(kron(eye(ny), D(nx-1,hx,2)) + kron(D(ny,hy,3),eye(nx - 1)))
Lv = eye((ny-1)*nx)/(dt) + mu*(kron(eye(ny-1),D(nx,hx,3))+kron(D(ny-1,hy,2),eye(nx)))
Lp = -(kron(eye(ny),D(nx,hx,1))+kron(D(ny,hy,1),eye(nx)))
Lp_coo = coo_matrix(Lp)
Lp_coo.data[0] = Lp_coo.data[0] - 1/hx**2
Lp_m = csr_matrix(Lp_coo)
 
```

```python
data_x = np.array([-np.ones(nx - 1), np.ones(nx - 1)])
offsets_x = [0, 1]
GX = kron(np.eye(ny),diags(data_x, offsets_x, shape=(nx - 1, nx))) / hx

# Gradient operator in y
data_y = np.array([-np.ones(nx * ny), np.ones(nx * ny)])
offsets_y = [0, nx]
GY = diags(data_y, offsets_y, shape=(nx * (ny - 1), nx * ny)) / hy

```

```python
def inter(A):
    A_shape = A.shape
    inter_A = np.zeros((A_shape[0]-1, A_shape[1]))
     
    for i in range(0,A_shape[0]-1):
        inter_A[i,:] = (A[i,:] + A[i+1,:])/2
    return inter_A

def diff(A):
    A_shape = A.shape
    diff_A = np.zeros((A_shape[0]-1, A_shape[1]))
     
    for i in range(0,A_shape[0]-1):
        diff_A[i,:] = A[i+1,:] - A[i,:]
    return diff_A
```

```python
%%time

# Boundary conditions
U = np.zeros((nx-1, ny))
V = np.zeros((nx, ny-1))

 
for t in np.arange(t_start, t_final+dt, dt):

    U_WE = np.concatenate((uW, U, uE), axis=0)
    U_all = np.concatenate((2*uS-U_WE[:,0].reshape(-1, 1), U_WE, 2*uN-U_WE[:,-1].reshape(-1, 1)), axis=1)
    

    V_SN = np.concatenate((vS, V, vN), axis=1) 
    V_all = np.concatenate((2*vW-V_SN[0,:], V_SN, 2*vE-V_SN[-1,:]), axis=0)

    Uc = inter(U_all[:,1:-1])
    Ux = diff(Uc)/hx
    Ua = np.transpose(inter(np.transpose(U_all[1:-1,:])))
    Uy = np.transpose(diff(np.transpose(Ua)))/hy
    
    Vc = np.transpose(inter(np.transpose(V_all[1:-1,:])))
    Vy = np.transpose(diff(np.transpose(Vc)))/hy
    Va = inter(V_all[:,1:-1])
    Vx = diff(Va)/hx
    
    Vb = inter(Vc)
    Ub = np.transpose(inter(np.transpose(Uc)))

    Uadvect = np.multiply(U,Ux) + np.multiply(Vb,Uy)
    Vadvect = np.multiply(Ub,Vx) + np.multiply(V,Vy)

    rhsu = (U / dt + Ubc - Uadvect).flatten(order='F')
    rhsv = (V / dt + Vbc - Vadvect).flatten(order='F')
     


    U = spsolve(Lu, rhsu)
    V = spsolve(Lv, rhsv)

    rhsp = (divbc-np.transpose(GX)*U-np.transpose(GY)*V)
    
    P = spsolve(Lp_m, rhsp)

    U = U - GX*P
    V = V - GY*P
     

    U = np.reshape(U,(nx-1,ny), order='F')
    V = np.reshape(V,(nx,ny-1), order='F')
    P = np.reshape(P,(nx,ny), order='F')
    
     
```

```python
# Subplot 1
plt.figure(figsize=(15, 12))

plt.subplot(2, 2, 1)
contour_U =  plt.contourf(x[1:-1], ym, U.T)
plt.axis([0, lx, 0, ly])
plt.colorbar(contour_U, label='U')

plt.subplot(2, 2, 2)
contour_V = plt.contourf(xm, y[1:-1], V.T)
plt.axis([0, lx, 0, ly])
plt.colorbar(contour_V, label='V')

plt.subplot(2, 2, 3)
contour_P = plt.contourf(xm, ym, P.T)
plt.axis([0, lx, 0, ly])
plt.colorbar(contour_P, label='P')


plt.subplot(2, 2, 4)
U_WE = np.transpose(inter(np.transpose(np.concatenate((uW, U, uE), axis=0))))
U_all = np.concatenate((uS, U_WE, uN), axis=1)
 

V_SN = inter(np.concatenate((vS, V, vN), axis=1) )
V_all = np.concatenate((vW, V_SN, vE), axis=0)
 

Len = np.sqrt(U_all**2+V_all**2+np.finfo(np.float64).eps)
plt.contourf(x, y, Len.T)
 
plt.colorbar()  # Add a color bar for reference
plt.quiver(x, y, (U_all/Len).T, (V_all/Len).T, scale=40, color='k')
# plt.streamplot(X, Y, U_all, V_all,  color='white')
# plt.show()
```
